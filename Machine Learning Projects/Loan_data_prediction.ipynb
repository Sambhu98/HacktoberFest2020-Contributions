```python
#importing necessary libraries
import pandas as pd
import numpy as np

#creating a datafrome of the train file
df=pd.read_csv('train.csv')
df.head(4)
```

```python
#checking the presence of null values in dataframe
df.isnull().sum()
```

```python
#dropping the records having NaN values in any of the features
df1=df.dropna()
df1.head(10)
```

```python
#Now let us check if we have cleaned data or not
df1.isnull().sum()
df1.head()
```

```python
#splitting the dataframe
target=df1['Loan_Status']
df=df1.drop(['Loan_Status'] ,axis='columns')
df.head()
```

```python
#importing necessaary mdels on which we will train the dataframe
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,f1_score
```

```python
c=df['Loan_ID'].tolist()
d=df['Gender'].tolist()
e=df['Married'].tolist()
f=df['Dependents'].tolist()
g=df['Education'].tolist()
h=df['Self_Employed'].tolist()
i=df['ApplicantIncome'].tolist()
j=df['CoapplicantIncome'].tolist()
k=df['LoanAmount'].tolist()
l=df['Loan_Amount_Term'].tolist()
m=df['Credit_History'].tolist()
n=df['Property_Area'].tolist()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Loan_ID']=le.fit_transform(c)
df['Gender']=le.fit_transform(df['Gender'])
df['Married']=le.fit_transform(e)
df['Dependents']=le.fit_transform(f)
df['Education']=le.fit_transform(g)
df['Self_Employed']=le.fit_transform(h)
df['ApplicantIncome']=le.fit_transform(i)
df['CoapplicantIncome']=le.fit_transform(j)
df['LoanAmount']=le.fit_transform(k)
df['Loan_Amount_Term']=le.fit_transform(l)
df['Credit_History']=le.fit_transform(m)
df['Property_Area']=le.fit_transform(n)
df.head()


#Loan_ID	Gender	Married	Dependents	Education	
#Self_Employed	ApplicantIncome	CoapplicantIncome	LoanAmount	Loan_Amount_Term	Credit_History	Property_Area
```

```python
#one hot encoding of the training dataframe
#df_final = pd.get_dummies(df) #done

#splitting the model in training and testing set
xtrain,xtest,ytrain,ytest=train_test_split(df,target,test_size=0.25,random_state=44)

#initializing the models
lr=LogisticRegression()
dt=DecisionTreeClassifier()
sv=SVC()
nb=GaussianNB()

#fitting the model with training and testing values
lr.fit(xtrain,ytrain)
dt.fit(xtrain,ytrain)
sv.fit(xtrain,ytrain)
nb.fit(xtrain,ytrain)

#predicting the values from the model
lr_pred=lr.predict(xtest)
dt_pred=dt.predict(xtest)
sv_pred=sv.predict(xtest)
nb_pred=nb.predict(xtest)

#comparing the accuracy score of the different models to select the best one
print('accuracy of logistic regression model=',accuracy_score(ytest,lr_pred))
print('accuracy of decision tree model=',accuracy_score(ytest,dt_pred))
print('accuracy of svm model=',accuracy_score(ytest,sv_pred))
print('accuracy of naive bayes model=',accuracy_score(ytest,nb_pred))
```
